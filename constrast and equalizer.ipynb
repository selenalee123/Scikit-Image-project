```python
from skimage import data,color,io
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray
import glob
from skimage.io import imread
from skimage import exposure
import matplotlib.pyplot as plt
```

```python
#import x_ray
example_file = glob.glob("chest_xray_image.png")[0]
chest_xray_image = imread(example_file)

# Show original x-ray image and its histogram
io.imshow(chest_xray_image)
```

```python
plt.title('Histogram of image')
plt.hist(chest_xray_image.ravel(), bins=120)
plt.show()

# Use histogram equalization to improve the contrast
xray_image_eq =  exposure.equalize_hist(chest_xray_image)

# Show the resulting image
io.imshow(xray_image_eq)
 
```

```python
#import map
image_aerial = glob.glob("map.png")[0]
image_aerial = imread(image_aerial)
io.imshow(image_aerial)
```

```python
# Use histogram equalization to improve the contrast
image_eq =  exposure.equalize_adapthist(image_aerial,clip_limit=0.03)

# Show the original and resulting image
io.imshow(image_aerial)
io.imshow(image_eq)
```
