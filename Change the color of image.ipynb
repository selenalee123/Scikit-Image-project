```python
# Import the modules from skimage
from skimage import data,color,io
import cv2
import numpy as np

```

```python
image = cv2.imread("rocket.png")
cv2.imshow("image",image)
cv2.waitKey(0)
rocket=data.rocket()
io.imshow(rocket)
```

```python
#convert the image to grayscale
gray_scaled_rocket  = color.rgb2gray(rocket)
io.imshow(gray_scaled_rocket)
```

```python
io.imshow(rocket)

io.imshow(gray_scaled_rocket)

#flip the image 
rocketvertical_flip=np.flipud(rocket)
io.imshow(rocketvertical_flip)



```
